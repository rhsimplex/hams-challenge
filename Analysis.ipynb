```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sample.csv')
```

```python
sns.pairplot(df[['para1', 'para2', 'para3', 'para4', 'price']])
```

```python
df = df[(df.para1 < 300)]
sns.pairplot(df, vars=['para1', 'para2', 'para3', 'para4', 'price'], hue='dow')
```

```python
t = df[df.loc2.isin(['03','04'])]
sns.pairplot(t, vars=['para1', 'para2', 'para3', 'para4', 'price'], hue='loc2')
```
